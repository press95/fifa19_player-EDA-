### The following columns need to be converted for further analysis:

'Value' Convert to Float and getting rid of currency symbol and suffix

'Wage' Convert to Float and rid of currency symbol and suffix.

'Joined' Convert to int with only year

'Contract Valid Until' Convert to datetime type

'Height' Convert to Float with decimal points

'Weight' Remove the suffix and convert to float

'Release Clause' Convert to Float and get rid of currency symbol and suffix.

```python
def strip_prefix_suffix(Value):
    
    if Value==np.nan:
        new_Value=float(0)
    else:
        Value=Value.replace('€','')
        if Value[-1]=='K':
            new_Value=float(Value.replace('K',''))
        else:
            new_Value=float(Value.replace('M',''))*1000
    return new_Value
```

```python
fifa_2['Value']= fifa_2['Value'].apply(lambda x : strip_prefix_suffix(x))
fifa_2['Wage']= fifa_2['Wage'].apply(lambda x : strip_prefix_suffix(x))
```

```python
fifa_2['Joined']= pd.to_datetime(fifa_2['Joined'], format='%b %d, %Y').apply(lambda x: x.year).fillna(0).astype(int)
```

```python
fifa_2['Contract Valid Until']= pd.to_datetime(fifa_2['Contract Valid Until'])
```

```python
fifa_2['Release Clause']= fifa_2['Release Clause'].fillna("0").apply(lambda x : strip_prefix_suffix(x))
```

```python
fifa_2['Height']= fifa_2['Height'].fillna("0'0").apply(lambda x: round(int(x.split("'")[1])/12+int(x.split("'")[0]),2))
```

```python
fifa_2['Weight']= fifa_2['Weight'].fillna('0').apply(lambda x: x.replace('lbs', '')).astype(float)
```

### whether Age affact over all rating ???

```python
def plot_scatter_plot(df,Nationality):
    good_p=df[(df.Nationality==Nationality)&(df.Overall>=80)]
    avg_p=df[(df.Nationality==Nationality)&(df.Overall<80)]
    plt.scatter(good_p.Age,good_p.Overall,color='blue',label='good_p',s=50)
    plt.scatter(avg_p.Age,avg_p.Overall,color='green',label='avg_p',marker='+',s=50)
    plt.xlabel('Age')
    plt.ylabel('Overall')
    plt.title('Nationality')
```

```python
plot_scatter_plot(Top_mid_df,'Spain')
```

As we can see above  "Age is justa number " where there are player's over all rating which is above 80.
